```python
from scipy.stats import linregress

# primero calculo el porcentaje de casos positivos por edad
positive_counts = lungcx_df[lungcx_df['LUNG_CANCER'] == 'YES']['AGE'].value_counts()
total_counts = lungcx_df['AGE'].value_counts()
positive_percent = (positive_counts / total_counts) * 100

# los distribuyo por edad
positive_percent = positive_percent.sort_index()

# Ploteo el gráfico
plt.figure(figsize=(10, 6))
plt.gcf().patch.set_facecolor('#5C5555')
plt.gca().set_facecolor('#5C5555')
bars = plt.bar(positive_percent.index, positive_percent.values, color='#BCF9A4')

# Realizo la linea de regresión
x = positive_percent.index.values
y = positive_percent.values
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_fit = slope * x + intercept

# Ploteo la linea de regresión
plt.plot(x, y_fit, color='#A60405', linewidth=2, label=f'Linear fit: y={slope:.2f}x + {intercept:.2f}')

# le agrego el "p-value"
plt.text(max(x), max(y_fit), f'p-value: {p_value:.6f}', ha='right', va='top', color='#A60405')


plt.title('Porcentaje de casos Positivos por Edad' , color='#F9F5F5')
plt.xlabel('Edad', color= '#F9F5F5')
plt.ylabel('Casos Positivos (%)', color='#F9F5F5')
plt.gca().tick_params(axis='x', colors='#F9F5F5')
plt.gca().tick_params(axis='y', colors='#F9F5F5')
plt.ylim(0, 100)
plt.legend()

plt.show()
```

Debido a que existe un "p-value" elevado (>0.05) podemos concluir que la variable independiente no genera ningun efecto sobre la variable dependiente, es decir los casos positivos de acuerdo con la edad. Sin embargo, no hay evidencia que sugiera que deba rechazarse la hipótesis nula.

*Since there is a high p-value we can conclude that the independent variable does not impact the dependent variable (ie. Positive Lung Cases to Age). Nevertheless, ther is no evidence att all that suggests the hypothesis should be rejected.*

*Devido ao fato de existir um valor de p elevado, podemos concluir que a variável independente não tem efeito sobre a variável dependente, ou seja, os casos positivos de acordo com a idade. Porém, não há evidência que determine que a hipótese deva ser rejeitada.*

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=50, shuffle=True)
```

Una vez entrenado el modelo con los datos de entrenamiento se realiza la prueba de predicción y luego se imprime en pantalla los resultados.

*Once the model is trained with the training data, the prediction test is performed, and then the results are displayed on the screen*.

*Uma vez treinado o modelo com os dados de treinamento, o teste de previsão é realizado e, em seguida, os resultados são exibidos na tela*.

```python
y_pred = log_reg.predict(x_test)
print(classification_report(y_pred=y_pred, y_true=y_test))
```
